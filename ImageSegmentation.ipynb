```python
def AndImage(a, b, T2):
    c = b.copy()
    for i in range(a.shape[0]):
        for j in range(a.shape[1]):
            if a[i][j] < T2 and b[i][j] == 255:
                c[i][j] = 150
            elif (a[i][j] >= T2 and b[i][j] == 255):
                c[i][j] = 255
    return c; 
```
